# Objective

#### There is a hotel chain which is having City Hotel and Resort Hotel which is facing high booking cancellation rate which lead to decrease in over all profitability.

#### Our objective is to analyse data and find out possible ways to increase the profitability



## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Visualizing coorelation using heat map

corr = df.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr, cmap='coolwarm', annot = True)
plt.show()
```

```python
# Rnaming hotel=1 by Resort Hotel and hotel=0 city hotel

df['hotel']= df['hotel'].replace(1,'Resort Hotel')
df['hotel']= df['hotel'].replace(0,'City Hotel')
df.head()

```

```python
df["hotel"]
```

```python
# Changing reservation_status_date type from object to date time

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
# Exploring unique values in each column

for col in df.describe(include = 'object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
# finding null values in each column

df.isnull().sum()
```

```python
# droping 'company' and 'agent' columns as its maximum values are null

df.drop(['company', 'agent'], axis = 1, inplace = True)
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Removing outlier from data frame

df = df[df['adr']<5000]
```

```python
# Replacing is_canceled=0 with Booking Not Cancelled and
# Replacing is_canceled=1 with Booking Cancelled

df['is_canceled']=df['is_canceled'].replace(0,"Booking Not Cancelled")

df['is_canceled']=df['is_canceled'].replace(1,"Booking  Cancelled")
```

```python
df
```

```python
# Finding month from data frame df

df['month'] = df['reservation_status_date'].dt.month
```

# Data Analysis and Visualizations

```python
# Visualizing cancelled vs not cancelled using bar plot

cancelled_perc = df['is_canceled'].value_counts()
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```
